```python
import os
import numpy as np
import pandas as pd
```

```python
# 設定 data_path
dir_data = './data/'
```

```python
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
```

### 如果沒有想法，可以先嘗試找出剛剛例子中提到的問題的答案
#### 資料的 row 數以及 column 數

```python
row = app_train.shape[0]
column = app_train.shape[1]
print('row : %d , column : %d'%(row,column))
app_train.sample(n=10)
```

#### 列出所有欄位

```python
for i in app_train.axes[1]:
    print(i)

for i in app_train.columns:
    print(i)
```

#### 截取部分資料

```python
dataofGenderIncome=app_train[['CODE_GENDER','AMT_INCOME_TOTAL']]
dataofGenderIncome.sample(n=10)
```
